написал функцию, которая вычисляет сумму MAU в столбцах Australia и Africa и проверяет, превышает ли эта сумма значение в столбце CIS.

Если Australia + Africa > CIS, то функция должна возвращать сумму Australia + Africa, в противном случае, должен возвращаться 0.

```python
def summ_(row_value):
    if (row_value['Australia'] + row_value['Africa']) > row_value['CIS']:
        return row_value['Australia'] + row_value['Africa']
    else:
        return 0
```

```python
mau_region['summ__'] = mau_region.apply(summ_, axis = 1)
mau_region
```

Создал новый столбец 'EU превышает порог'. Он должен принимать значение True, в ситуациях, когда EU > 4000 и False в противном случае.

### Функция исследования DAU

Написал обобщенную функцию исследования DAU. Функция должна принимать на вход имя файла с данными DAU. И выводить на экран:

Таблицу DAU;

Линейный график DAU;

График долей пользователей (область с накоплением).

использовал цикл по столбцам для подсчета долей пользователей. Не забыл убрать столбец Total перед отрисовкой графика.

```python
def dau_research(path:str):
    dau = pd.read_excel(path)
    dau['date'] = pd.to_datetime(dau['date'])
    dau = dau.set_index('date')
    
    # построение абсолютного графика
    dau.plot(figsize=(10,5), grid=True)
    plt.title('Линейный график'), plt.xlabel('Date'), plt.ylabel('Пользователи')
    plt.show()
    
    # построение относительного графика
    dau['total'] = 0
    for itm in dau.columns:
        dau['total'] += dau[itm]
        
        
    for itm in dau.columns[:-1]:
        dau[itm] =  dau[itm] / dau['total']
        
    dau.drop(columns = 'total').plot.area(figsize = (15, 5))
    plt.title('Изменение DAU'), plt.xlabel('Date'), plt.ylabel('Процент уникальных пользователей')
    plt.show()
        
    return dau
```

```python
dau_research('dau_channel_.xlsx')
```
